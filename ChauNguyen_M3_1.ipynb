3. Re-compute $ (1 - f) \otimes h$, with $h$ and $f$ as specified above, but using this new `my_array_ref` function.

4. Compare the output obtained in step 3 above with that in step 1.

```python
import numpy as np

#--------------------------------------------------------
# Set up image f and kernel h
h = np.array([[-1, 1],[-1, 1]])
f = np.zeros((10,10))
f[3:8,3:8] = np.ones((5, 5)) #block assignment of values

print("Image f=\n", f, "\nFilter/Kernel h=\n", h)

#--------------------------------------------------------
# Returning zero when matrix indices are out of bounds
def my_array_ref (a, i, j):
    if ((i >= 0) & (i < a.shape [0]) & (j >= 0) & (j < a.shape [1])):
        return a[i, j]
    else :
        return 0

# Returning boundary value when matrix indices are out of bounds
def new_my_array_ref (a, i, j):
    if (i < 0) or (i >= a.shape[0]):
      i = max(0, min(i, a.shape[0] - 1))
    if (j < 0) or (j >= a.shape[1]):
      j = max(0, min(j, a.shape[1] - 1))
    return a[i, j]

#--------------------------------------------------------
# Compute step 1 ((1 - f) * h with old my_array_ref function)
new_f = 1 - f
g1 = np.zeros((10, 10)) #output array
for i in range(new_f.shape[0]) :
    for j in range (new_f.shape[1]):
        prod_sum = 0.0
        for k in range (h.shape[0]) :
            for l in range (h.shape[1]) :
                prod_sum = prod_sum + my_array_ref(h, k, l) * my_array_ref(new_f, i+k, j+l)
        g1[i, j] = prod_sum
print("\nStep 1 (old my_array_ref): Output computed by (1-f) * h=\n", g1)

# Compute step 3 ((1 - f) * h with new my_array_ref function)
g2 = np.zeros((10, 10)) #output array
for i in range(new_f.shape[0]) :
    for j in range (new_f.shape[1]):
        prod_sum = 0.0
        for k in range (h.shape[0]) :
            for l in range (h.shape[1]) :
                prod_sum = prod_sum + new_my_array_ref(h, k, l) * new_my_array_ref(new_f, i+k, j+l)
        g2[i, j] = prod_sum
print("\nStep 3 (new my_array_ref): Output computed by (1-f) * h=\n", g2)
```

## Assignment: Difference of Gaussian (DoG)

Interestingly, the second derivative of a Gaussian has a very similar shape to the difference of two Gaussians (DoG) of different scales, $h_{\sigma_1}(x) - h_{\sigma_2}(x)$.

Find values of $\sigma_1$ and $\sigma_2$ for which the DoG is most similar to the second derivative of Gaussian of $\sigma = 3$. Plot the functions.

```python
import matplotlib.pyplot as plt
from scipy.optimize import minimize

#------------------------------------------------------------------
# FUNCTION DEFINITIONS
def gaussian (x, sigma):
    return (np.exp(-(x**2/(2*sigma*sigma)))*(1/(np.sqrt(2*np.pi)*sigma)))

def second_derivative_of_gaussian (x, sigma):
    return (np.exp(-(x**2/(2*sigma*sigma)))*(-(1 - x*x/(sigma*sigma))/(np.sqrt(2*np.pi)*sigma**3)))

def difference_of_gaussian (x, sigma_1, sigma_2):
    return gaussian(x, sigma_1) - gaussian(x, sigma_2)

def objective_function(sigmas, x):
    sigma1, sigma2 = sigmas
    dog = difference_of_gaussian(x, sigma1, sigma2)
    second_derivative = second_derivative_of_gaussian(x, sigma=3)
    difference = np.sum(np.abs(dog - second_derivative))
    return difference

#------------------------------------------------------------------
# Function calls

t = np.linspace(-10, 10, 101)
second_derivative = second_derivative_of_gaussian(t, 3)

# Initial guess for sigmas
initial_sigmas = [1, 2]

# Optimize to find the best values for sigmas
result = minimize(objective_function, initial_sigmas, args=(t,), bounds=[(1, 5), (1, 5)])

# Extract the optimal sigmas
best_sigma1, best_sigma2 = result.x

#------------------------------------------------------------------
# Function plots for comparison
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.plot(t, second_derivative, label='Second Derivative (sigma=3)')
plt.plot(t, difference_of_gaussian(t, best_sigma1, best_sigma2), label=f'DoG (sigma1={best_sigma1:.2f}, sigma2={best_sigma2:.2f})')
plt.title('Comparison of Second Derivative of Gaussian and DoG')
plt.legend()

plt.subplot(1,3,2)
plt.plot(t, second_derivative, label='Second Derivative (sigma=3)')
plt.title('Second Derivative of Gaussian (sigma=3)')
plt.legend()

plt.subplot(1,3,3)
plt.plot(t, difference_of_gaussian(t, best_sigma1, best_sigma2), label=f'DoG (sigma1={best_sigma1:.2f}, sigma2={best_sigma2:.2f})')
plt.title('Difference of Gaussian')
plt.legend()

print(f"Optimal values: sigma_1 = {best_sigma1:.2f}, sigma_2 = {best_sigma2:.2f}")
```
